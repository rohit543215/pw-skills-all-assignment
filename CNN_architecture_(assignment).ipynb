## Q1. What is the role of the filters and feature maps in convolutional neural network(CNN)?

## Filter
A filter is a small matric (e.g, 3*3) that slides over the input image pixel by pixel, scanning small regions at a time to detect specific pattern like edges, corners, or textures.

## Feature map
A feature mmpa is the output produced after a filter finishes scanning the entire image. it is a 2D grid that stores the results of the filter's scan, showing where a particular pattern was detected in the image.

Each filter produces one feature map. so if a convolutional layer has 32 filters, it produces 32 featuers map each one  representing a diff pattern detected from the input. in early layers, filters detect simple features like edges and corners. in deeper layers, they detect complex features like shape and objects.
This process of filters extracting
pattern and storing them as feature maps is what allow CNN's to automatically learn and recognize visual features from images.

## Q2.  Explain the concepts of padding and stride in CNNs(Convolutional Neural Network). How do they affect the output dimensions of feature maps?

## Padding:
padding is the process of adding  extra pixels around the borderof the input image before  applying the filter. when  a filter slides over on image, the edge and corner pixels get scanned less compared to to the center pixels, causing loss of edge information padding solves this by bringing the  edge pixels more towards the center, ensuring they control the size of the output feature map. more padding results in a longer output size.

## Stride:
stride is the number of pixels the  filter moves at each step while sliding over the image. a stride of 1 pixel means the filter moves one pixel at a time,scanning every detrail slowly and  producing a larger feature map. a stride of 2 means the filter jumps 2 pixels at a time, which is faster but some information is skipped, producing a smaller featuer map.
## Effect on output dimensions.
- more padding = output feature map size increases.
- larger stride = output feature map size decreases.
- smaller stride = output feature map increases.
- no padding + larger stride = output shrinks significantly

in practice, padding is used to preserve the spatial information and maintain feature map size, while stride is used to reduce dimensions and computational cost. both together give control over how the network precesses and compresses image information layer by layer.

## Q3. Define receptive field in the context of CNN's. why is  it important for  deep architecture?


## Receptive field in CNN's
the receptive field refers to the region of the input image that a particular filter neuron can 'see' at a time. in a single convolutional layer with a 3*3 pixels, a very small portion of the image.

However, as the network gets deeper, the receptive field grows. each deeper layer's filter looks at the output of the previous layer, which itself was already looking at a region of the original image. so indirectly, neurons in deeper layers are seeing a much larger portion of th original input than neurons in early layers.

## why it is important for deep architectures:
In early layers, the small receptive field allows the network to combine those simple features and detect complex, large-scale pattern like faces, cars, or  entire objects. which are spread across large regions of the image.
This is exactly why deep architectures are powerful. stacking multiple layers fields, allowing the network to understand both fine-grained local details and high level global structure of the image without a sufficiently large receptive field. the network would never be able to recognize complex objects that span large portions of the input.

## Q4. Discuss how filter size and stride infuence the number of parameters in a CNN.

## Parameters in a convolutional layer.
- The no of parameters in a convolutional layer is determined by:
(filter_h * filter_w * input_channels+1)* num_filters
- The (+1) accounts for the bias term of each filter.

## Effect of filter size
- filter size directly affects the number of parameters. a larger filter has more weights, so the total parameter count  increases.

- A 5*5 filter gives nearly 3x more parameters than a 3*3 filter. this  is why modern  architecutures like VGG prefer stacking two 3*3 layers over one 5*5 same receptive field, fewer parameters.
## Effect of stride
- stride gas bi  effect on the no of parameters. ir only  controls the output feature map size.

## Q8. Using PyTorch, write a script to define and train a CNN on the MNIST dataset. Include model definition, data loaders, training loop, and accuracy evaluation.

````python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset, random_split
import pandas as pd
import numpy as np
import os

# =========================================
# 📂 LOAD DATA
# =========================================
train_df = pd.read_csv("/content/digit-recognizer-data/train.csv")
test_df  = pd.read_csv("/content/digit-recognizer-data/test.csv")

y_train = train_df['label'].values                  # labels (42000,)
x_train = train_df.drop('label', axis=1).values     # pixels (42000, 784)
x_test  = test_df.values                            # pixels (28000, 784)

print(f"Train samples : {x_train.shape[0]}")
print(f"Test  samples : {x_test.shape[0]}")

# =========================================
# 🔄 PREPROCESS
# =========================================
# Reshape to (N, 1, 28, 28) — PyTorch uses (batch, channels, height, width)
# TensorFlow used (batch, height, width, channels) — note the difference
x_train = x_train.reshape(-1, 1, 28, 28) / 255.0
x_test  = x_test.reshape(-1, 1, 28, 28)  / 255.0

# Convert to PyTorch tensors
# x → float32 (model weights are float32)
# y → long (int64) — PyTorch CrossEntropyLoss expects long, not one-hot
x_train_tensor = torch.tensor(x_train, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train, dtype=torch.long)
x_test_tensor  = torch.tensor(x_test,  dtype=torch.float32)

# NOTE: No to_categorical() needed in PyTorch
# PyTorch's CrossEntropyLoss handles raw integer labels internally

# =========================================
# ✂ TRAIN / VALIDATION SPLIT (80/20)
# =========================================
dataset = TensorDataset(x_train_tensor, y_train_tensor)

train_size = int(0.8 * len(dataset))
val_size   = len(dataset) - train_size

train_dataset, val_dataset = random_split(dataset, [train_size, val_size])

# DataLoader batches and shuffles data automatically
train_loader = DataLoader(train_dataset, batch_size=86, shuffle=True)
val_loader   = DataLoader(val_dataset,   batch_size=86, shuffle=False)

# =========================================
# 🤖 MODEL
# =========================================
class DigitCNN(nn.Module):
    """
    CNN for digit recognition.
    Architecture mirrors the Keras version:
    Conv→Conv→Pool→Dropout→Conv→Conv→Pool→Dropout→Flatten→Dense→Dropout→Output
    """
    def __init__(self):
        super(DigitCNN, self).__init__()

        self.features = nn.Sequential(
            # Block 1 — 32 filters, 5x5 kernel
            nn.Conv2d(1, 32, kernel_size=5, padding='same'),   # (N, 32, 28, 28)
            nn.ReLU(),
            nn.Conv2d(32, 32, kernel_size=5, padding='same'),  # (N, 32, 28, 28)
            nn.ReLU(),
            nn.MaxPool2d(2, 2),                                 # (N, 32, 14, 14)
            nn.Dropout(0.25),

            # Block 2 — 64 filters, 3x3 kernel
            nn.Conv2d(32, 64, kernel_size=3, padding='same'),  # (N, 64, 14, 14)
            nn.ReLU(),
            nn.Conv2d(64, 64, kernel_size=3, padding='same'),  # (N, 64, 14, 14)
            nn.ReLU(),
            nn.MaxPool2d(2, 2),                                 # (N, 64, 7, 7)
            nn.Dropout(0.25),
        )

        self.classifier = nn.Sequential(
            nn.Flatten(),           # (N, 64*7*7) = (N, 3136)
            nn.Linear(3136, 256),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(256, 10)
            # No softmax here — CrossEntropyLoss applies it internally
        )

    def forward(self, x):
        x = self.features(x)
        x = self.classifier(x)
        return x

# =========================================
# ⚙ SETUP
# =========================================
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

model     = DigitCNN().to(device)
optimizer = optim.RMSprop(model.parameters(), lr=0.001, alpha=0.9, eps=1e-08)
criterion = nn.CrossEntropyLoss()
# alpha in PyTorch RMSprop = rho in Keras RMSprop — same thing

# =========================================
# 📉 CALLBACKS (manual in PyTorch)
# =========================================
# PyTorch has no built-in callbacks — we implement them manually

best_val_acc   = 0.0
patience       = 5      # EarlyStopping patience
no_improve     = 0      # counter
lr_patience    = 3      # ReduceLROnPlateau patience
lr_no_improve  = 0
min_lr         = 1e-5

# =========================================
# 🏋 TRAINING LOOP
# =========================================
epochs = 3

for epoch in range(epochs):
    # ── TRAIN ──────────────────────────────
    model.train()   # enables Dropout
    train_loss = 0
    train_correct = 0

    for x_batch, y_batch in train_loader:
        x_batch = x_batch.to(device)
        y_batch = y_batch.to(device)

        optimizer.zero_grad()           # clear old gradients
        outputs = model(x_batch)        # forward pass
        loss = criterion(outputs, y_batch)  # compute loss
        loss.backward()                 # backprop
        optimizer.step()                # update weights

        train_loss    += loss.item()
        train_correct += (outputs.argmax(1) == y_batch).sum().item()

    train_acc = train_correct / train_size

    # ── VALIDATE ───────────────────────────
    model.eval()    # disables Dropout
    val_loss = 0
    val_correct = 0

    with torch.no_grad():   # no gradient computation during eval
        for x_batch, y_batch in val_loader:
            x_batch = x_batch.to(device)
            y_batch = y_batch.to(device)

            outputs  = model(x_batch)
            loss     = criterion(outputs, y_batch)
            val_loss += loss.item()
            val_correct += (outputs.argmax(1) == y_batch).sum().item()

    val_acc = val_correct / val_size

    print(f"Epoch {epoch+1}/{epochs} "
          f"— train_acc: {train_acc:.4f} "
          f"— val_acc: {val_acc:.4f}")

    # ── SAVE BEST MODEL (ModelCheckpoint) ──
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        torch.save(model.state_dict(), "best_model.pt")
        print(f"  ✅ Best model saved (val_acc={val_acc:.4f})")
        no_improve    = 0
        lr_no_improve = 0
    else:
        no_improve    += 1
        lr_no_improve += 1

    # ── REDUCE LR ON PLATEAU ───────────────
    if lr_no_improve >= lr_patience:
        current_lr = optimizer.param_groups[0]['lr']
        new_lr     = max(current_lr * 0.5, min_lr)
        for pg in optimizer.param_groups:
            pg['lr'] = new_lr
        print(f"  📉 LR reduced: {current_lr:.6f} → {new_lr:.6f}")
        lr_no_improve = 0

    # ── EARLY STOPPING ─────────────────────
    if no_improve >= patience:
        print(f"  ⏹ Early stopping at epoch {epoch+1}")
        break

# =========================================
# 🔮 PREDICT ON KAGGLE TEST DATA
# =========================================
# Load best weights back
model.load_state_dict(torch.load("best_model.pt"))
model.eval()

all_preds = []

with torch.no_grad():
    for i in range(0, len(x_test_tensor), 86):
        batch   = x_test_tensor[i:i+86].to(device)
        outputs = model(batch)
        preds   = outputs.argmax(1).cpu().numpy()
        all_preds.extend(preds)

pred_labels = np.array(all_preds)

# =========================================
# 📤 SUBMISSION
# =========================================
submission = pd.DataFrame({
    "ImageId": range(1, len(pred_labels) + 1),
    "Label":   pred_labels
})

submission.to_csv("submission.csv", index=False)
print(os.listdir("."))
print(submission.head())
````
