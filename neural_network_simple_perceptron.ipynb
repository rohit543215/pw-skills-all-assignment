#Q2. Explain the basic architecture and functioning of a Perceptron. What are its limitations?

## Introduction

A Perceptron is one of the earliest and simplest types of artificial neural networks. It was introduced by Frank Rosenblatt in 1958. It is mainly used for binary classification problems (two-class problems).

It mimics a biological neuron and forms the foundation of modern deep learning models.

## Basic architecture of a perceptron

A perceptron consists of:

- 1.Input Layer (x₁, x₂, x₃, …, xₙ)
These are input features.

- 2.Weights (w₁, w₂, w₃, …, wₙ)
Each input has a weight that represents its importance.

- 3.Bias (b)
A constant value added to adjust the output.

- 4.Summation Function
It calculates the weighted sum

𝑧=(𝑤1𝑥1+𝑤2𝑥2+...+𝑤𝑛𝑥𝑛)+𝑏

- 5.Activation Function (Step Function)
If z ≥ 0 → output = 1
If z < 0 → output = 0

So, the perceptron produces a binary output (0 or 1).

## Working of a Perceptron

Step-by-step process:

- 1.Inputs are multiplied by their respective weights.

- 2.The weighted sum is calculated.

- 3.Bias is added.

- 4.The result passes through a step activation function.

- 5.Final output is either 0 or 1.

## Training Process

The perceptron learns by adjusting weights using this rule:

𝑤=𝑤+𝜂(𝑦−𝑦^)𝑥

Where:

η = learning rate

y = actual output

ŷ = predicted output

If prediction is wrong, weights are updated to reduce error.

## Limitations of Perceptron

- 1.Can only solve linearly separable problems
It cannot solve non-linear problems like XOR.

- 2.Single-layer structure
It has no hidden layers, so it cannot learn complex patterns.

- 3.Binary output only
It performs only two-class classification.

- 4.Uses step activation function
The step function is not differentiable, limiting advanced optimization.

# Q3. Describe the purpose of activation function in neural networks. Compare Sigmoid, ReLU, and Tanh functions.

## Purpose of Activation Function

An activation function determines whether a neuron should be activated or not by transforming the weighted sum of inputs into an output signal.

In simple words, it introduces non-linearity into the neural network.

## Why is it important?

- Without activation functions, a neural network would behave like a linear model, no matter how many layers it has.

- It helps the network learn complex patterns.

- It controls the output range of a neuron.

- It allows deep networks to solve non-linear problems.

## 1. Sigmoid Function
𝜎(𝑥)=1/1+𝑒^−𝑥

	​Range: (0, 1)
## Characteristics:

- Used mainly for binary classification.
- Output can be interpreted as probability.
- Smooth and differentiable.

##Advantages:

- Good for output layer in binary classification.

## Disadvantages:
- Suffers from vanishing gradient problem.
- Outputs are not zero-centered.
- Slower convergence.



## 2. Tanh (Hyperbolic Tangent)

tanh(x)=e^x+e^−x / ex−e−x
	​

    Range: (-1, 1)

## Characteristics:

- Zero-centered output.
- Stronger gradients than sigmoid.

## Advantages:

- Performs better than sigmoid in hidden layers.
- Zero-centered output improves optimization.

## Disadvantages:
- Still suffers from vanishing gradient for large values.




## 3. ReLU (Rectified Linear Unit)

ReLU(x)=max(0,x)

    Range: [0, ∞)

## Characteristics:
- Returns 0 for negative values.
- Linear for positive values.

## Advantages:
- Computationally efficient.
- Reduces vanishing gradient problem.
- Faster training.
- Most widely used in deep networks.

## Disadvantages:
- Can cause “Dead Neuron Problem” (neurons stuck at 0).

| Feature            | Sigmoid               | Tanh                         | ReLU                                 |
| ------------------ | --------------------- | ---------------------------- | ------------------------------------ |
| Output Range       | (0, 1)                | (-1, 1)                      | [0, ∞)                               |
| Zero-centered      | No                    | Yes                          | No                                   |
| Vanishing Gradient | Yes                   | Yes (less than sigmoid)      | Rare                                 |
| Training Speed     | Slow                  | Moderate                     | Fast                                 |
| Common Usage       | Output layer (binary) | Hidden layers (older models) | Hidden layers (modern deep learning) |


# Q.4 What is the difference between Loss function and Cost function in neural networks? Provide examples.

## 1. Loss Function

A loss function measures the error for a single training example.
It tells us how far the predicted output is from the actual output for one data point.

In simple words:

Loss = Error for one sample.

### Example:

If the actual value is 10 and the predicted value is 8,

Loss = 10 − 8 = 2 (depending on the formula used)

### Common Loss Functions:
- Mean Squared Error (MSE) (per sample):
    - Loss=(y−ypred)^2

- Binary Cross-Entropy (for one sample):

    - Loss=−[ylog(ypred)+(1−y)log(1−ypred)]

## 2. Cost Function

A cost function is the average (or total) loss over the entire training dataset.

In simple words:
- Cost = Average error of all samples.

If we have N training samples:

𝐶𝑜𝑠𝑡=1/𝑁∑loss(i)

So, the cost function measures the overall performance of the model.

| Aspect   | Loss Function                  | Cost Function                     |
| -------- | ------------------------------ | --------------------------------- |
| Measures | Error for one training example | Average error over entire dataset |
| Scope    | Single data point              | Entire dataset                    |
| Purpose  | Computes individual error      | Used for optimization             |
| Used In  | Calculating sample error       | Training the model                |


##  Example

Suppose we have 3 predictions:
| Actual (y) | Predicted (ŷ) | Loss (MSE) |
| ---------- | ------------- | ---------- |
| 5          | 4             | (5−4)² = 1 |
| 7          | 6             | (7−6)² = 1 |
| 9          | 8             | (9−8)² = 1 |


Total Cost (MSE) = (1 + 1 + 1) / 3 = 1

- Here:
  - Each squared error = Loss
  - Average of all = Cost

# Q5. What is the role of optimizers in neural networks? Compare Gradient Descent, Adam, and RMSprop.  

## Role of Optimizers in Neural Networks

An optimizer is an algorithm used to adjust the weights and biases of a neural network in order to minimize the loss (or cost) function.

In simple words:

- Optimizers help the model learn by reducing error step by step.

During training:

- 1.The model makes predictions.
- 2.The loss function calculates the error.
- 3.The optimizer updates the weights using gradients.
- 4.This process repeats until the error becomes small.

Most optimizers are based on Gradient Descent, which updates weights using:

## 1. Gradient Descent

Gradient Descent updates weights using the gradient of the loss function.
### Types:
- Batch Gradient Descent
- Stochastic Gradient Descent (SGD)
- Mini-batch Gradient Descent

### Advantages:
- Simple and easy to implement.
- Works well for smaller datasets.

### Disadvantages:
- Slow convergence.
- Can get stuck in local minima.
- Sensitive to learning rate.
- May oscillate in steep regions.



## 2.RMSprop (Root Mean Square Propagation)
RMSprop improves Gradient Descent by adjusting the learning rate for each parameter individually.

It:

- Divides the gradient by a moving average of squared gradients.
- Reduces oscillations.
- Works well for non-stationary problems.

### Advantages:

- Faster convergence than SGD.
- Handles varying gradients better.

### Disadvantages:
- Still requires tuning learning rate.



## 3. Adam (Adaptive Moment Estimation)

Adam combines:
- Momentum (like storing past gradients)
- RMSprop (adaptive learning rates)

It keeps track of:
- First moment (mean of gradients)
- Second moment (variance of gradients)

### Advantages:
- Fast convergence.
- Adaptive learning rate.
- Works well in most practical problems.
- Less tuning required.

### Disadvantages:
- Can sometimes overfit.
- Slightly more complex.

| Handles Oscillation | No               | Yes      | Yes                      |

| Common Usage        | Basic problems   | RNNs     | Most deep learning tasks |


## Q6.Write a Python program to implement a single-layer perceptron from scratch using NumPy to solve the logical AND gate. give code in one go

```python
import numpy as np

# Step 1: Define AND gate dataset
# Inputs
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

# Outputs
y = np.array([0, 0, 0, 1])

# Step 2: Initialize parameters
np.random.seed(42)
weights = np.random.randn(2)
bias = np.random.randn()

learning_rate = 0.1
epochs = 10

# Step 3: Define activation function (Step Function)
def step_function(x):
    return 1 if x >= 0 else 0

# Step 4: Training
for epoch in range(epochs):
    print(f"\nEpoch {epoch+1}")
    for i in range(len(X)):
        # Linear combination
        linear_output = np.dot(X[i], weights) + bias

        # Prediction
        y_pred = step_function(linear_output)

        # Error
        error = y[i] - y_pred

        # Update weights and bias
        weights = weights + learning_rate * error * X[i]
        bias = bias + learning_rate * error

        print(f"Input: {X[i]}, Predicted: {y_pred}, Actual: {y[i]}, Error: {error}")

# Step 5: Final Results
print("\nFinal Weights:", weights)
print("Final Bias:", bias)

print("\nTesting Trained Perceptron:")
for i in range(len(X)):
    linear_output = np.dot(X[i], weights) + bias
    y_pred = step_function(linear_output)
    print(f"Input: {X[i]}, Output: {y_pred}")
```

## Q7.Implement and visualize Sigmoid, ReLU, and Tanh activation functions using Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate input values
x = np.linspace(-10, 10, 400)

# Define activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

# Plot Sigmoid
plt.figure()
plt.plot(x, sigmoid(x))
plt.title("Sigmoid Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.show()

# Plot ReLU
plt.figure()
plt.plot(x, relu(x))
plt.title("ReLU Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.show()

# Plot Tanh
plt.figure()
plt.plot(x, tanh(x))
plt.title("Tanh Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.show()
```

## Q8: Use Keras to build and train a simple multilayer neural network on the MNIST digits dataset. Print the training accuracy.

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Load MNIST dataset
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

# Normalize the data (0-255 → 0-1)
x_train = x_train / 255.0
x_test = x_test / 255.0

# Build the model (Multilayer Perceptron)
model = keras.Sequential([
    layers.Flatten(input_shape=(28, 28)),   # Convert 28x28 image to 1D vector
    layers.Dense(128, activation='relu'),   # Hidden layer
    layers.Dense(64, activation='relu'),    # Hidden layer
    layers.Dense(10, activation='softmax')  # Output layer (10 digits)
])

# Compile the model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Train the model
history = model.fit(x_train, y_train, epochs=5, batch_size=32, verbose=1)

# Print training accuracy
train_loss, train_accuracy = model.evaluate(x_train, y_train, verbose=0)
print("\nTraining Accuracy:", train_accuracy)
```

## Q.9: Visualize the loss and accuracy curves for a neural network model trained on the Fashion MNIST dataset. Interpret the training behavior.

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt
import numpy as np

# Load Fashion MNIST dataset
(x_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()

# Normalize data
x_train = x_train / 255.0
x_test = x_test / 255.0

# Build neural network
model = keras.Sequential([
    layers.Flatten(input_shape=(28, 28)),
    layers.Dense(128, activation='relu'),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# Compile model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Train model
history = model.fit(
    x_train, y_train,
    epochs=5,
    batch_size=32,
    validation_data=(x_test, y_test)
)

# Plot Training & Validation Loss
plt.figure()
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title("Training and Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend(["Training Loss", "Validation Loss"])
plt.grid(True)
plt.show()

# Plot Training & Validation Accuracy
plt.figure()
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title("Training and Validation Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend(["Training Accuracy", "Validation Accuracy"])
plt.grid(True)
plt.show()
```

Question 10: You are working on a project for a bank that wants to automatically detect fraudulent transactions. The dataset is large, imbalanced, and contains structured features like transaction amount, merchant ID, and customer location. The goal is to classify each transaction as fraudulent or legitimate. Explain your real-time data science workflow:

 ● How would you design a deep learning model (perceptron or multilayer NN)?

 ● Which activation function and loss function would you use, and why?

 ● How would you train and evaluate the model, considering class imbalance?

 ● Which optimizer would be suitable, and how would you prevent overfitting?


## **1️⃣ Designing the Deep Learning Model**

Since the dataset:

* Is **large**
* Is **imbalanced**
* Contains **structured/tabular features**

A **Multilayer Neural Network (MLP)** is more suitable than a single perceptron.

### Why not a single perceptron?

* A perceptron is a linear classifier.
* Fraud detection patterns are usually **non-linear and complex**.

### Proposed Model Architecture:

* **Input Layer** → Number of neurons = number of features
* **Hidden Layers** → 2–3 dense layers (e.g., 128 → 64 → 32 neurons)
* **Output Layer** → 1 neuron (Binary classification: Fraud / Legitimate)

Example structure:

Input → Dense(128) → Dense(64) → Dense(32) → Output(1)

This allows the model to learn complex fraud patterns.

---

## **2️⃣ Activation Function & Loss Function**

### 🔹 Hidden Layers Activation: **ReLU**

* Computationally efficient
* Avoids vanishing gradient problem
* Works well for structured tabular data

### 🔹 Output Layer Activation: **Sigmoid**

* Outputs probability between 0 and 1
* Perfect for binary classification

[
P(y=1) = \frac{1}{1 + e^{-z}}
]

### 🔹 Loss Function: **Binary Cross-Entropy**

[
Loss = -[y \log(\hat{y}) + (1-y)\log(1-\hat{y})]
]

Why?

* Suitable for binary classification
* Works well with sigmoid output
* Handles probability-based predictions

---

## **3️⃣ Training & Evaluation (Handling Class Imbalance)**

Since fraud cases are rare:

### Problem:

* Model may predict all transactions as legitimate.
* High accuracy but poor fraud detection.

### Solutions:

#### ✅ 1. Use Class Weights

Assign higher weight to fraud class during training.

Example:

* Legitimate = weight 1
* Fraud = weight 10 (or higher)

This penalizes wrong fraud predictions more.

#### ✅ 2. Resampling Techniques

* Oversampling minority class (SMOTE)
* Undersampling majority class

#### ✅ 3. Use Proper Evaluation Metrics

Accuracy is NOT suitable.

Use:

* Precision
* Recall
* F1-Score
* ROC-AUC
* PR-AUC (very important for imbalanced data)

Focus especially on:

* **Recall for fraud class** (catch as many fraud cases as possible)

---

## **4️⃣ Optimizer Selection**

### Best Choice: **Adam**

Why?

* Adaptive learning rate
* Fast convergence
* Works well for large datasets
* Handles sparse gradients efficiently

Adam combines:

* Momentum
* RMSprop

---

## **5️⃣ Preventing Overfitting**

Fraud detection models can overfit due to:

* High dimensional features
* Noise in data

### Techniques:

#### 🔹 Dropout (e.g., 0.3–0.5)

Randomly disables neurons during training.

#### 🔹 L2 Regularization

Adds penalty to large weights.

#### 🔹 Early Stopping

Stop training when validation loss stops improving.

#### 🔹 Batch Normalization

Stabilizes training and improves generalization.

---

# **6️⃣ Real-Time Deployment Considerations**

For real-time fraud detection:

* Model must be fast (low latency)
* Use probability threshold tuning (not fixed 0.5)
* Continuously retrain with new fraud patterns
* Monitor model drift

---

# **Final Summary**

For fraud detection:

* Use **Multilayer Neural Network**
